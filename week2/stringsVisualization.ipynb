# Generate a time series

```python
import numpy as np
radians = np.arange(0, 2*np.pi, 0.01);
```

# Plot the time series using the matplotlib library

```python
import matplotlib.pylab as plt
# plot the time series against the same time series
# https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html
# plt.plot(xvals, yvals)
plt.plot(radians, radians)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('tight')
plt.show()
```

# Plot the time series against the sine of the time series

```python
sine = np.sin(radians)
sine
```

```python
plt.plot(radians, sine)
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.show()
```

# Create strings for the x and y axis labels

```python
xlabel = 'Radians'
ylabel = 'Amplitude'
```

# Plot using the label strings

```python
plt.plot(radians, np.sin(radians))
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.show()
```

# Plot the sine wave with degrees in the x-axis instead of radians

```python
# use the numpy function rad2deg to convert
xlabel = 'Degrees'
plt.plot(np.rad2deg(radians), np.sin(radians))
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.show()
```

# Import an audio file from class GitHub repo using librosa library

```python
!git clone https://github.com/jcdevaney/imc2023.git

import librosa
import librosa.display

file = '/content/imc2023/audioFiles/avm.wav'
sig , sr = librosa.load(file,mono=True,sr=None)
```

# Calculate the length in seconds

```python
length_s = sig.shape[0] / sr
print(length_s)

print(sig.shape[0])
print(sr)
```

# Plot the time-domain representation of avm.wav

```python
xlabel = 'Samples'
plt.plot(sig)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
```

# Plot with time on the x-axis

```python
# divide by the sampling rate
time=np.arange(0,sig.shape[0])/sr
xlabel = 'Times (Seconds)'
plt.plot(time,sig)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
print(time)
```

# Plot with amplitude on the y-axis

```python
# plot the absolute value of the signal
plt.plot(time,abs(sig))
plt.xlabel(xlabel)
plt.ylabel('Absolute Amplitude')

# find the maximum value in the signal
excursion=max(abs(sig))
print(excursion)
```

```python
# create a new signal by dividing the
# original signal by the maximum value
# this will scale the signal to be between -1 and 1
sigNew = sig / excursion
plt.plot(time,sigNew)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
```
